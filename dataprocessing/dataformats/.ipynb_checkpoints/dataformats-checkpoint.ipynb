```python
import pandas as pd
import numpy as np
from numpy.random import randint
from numpy.random import random_sample

import sagemaker.amazon.common as smac
```

```python
x1 = randint(0,100,size=100) #100 numbers b/w 1 and 100
x2 = random_sample(size=100) #100 numbers b/w [0 and 1)
x3 = random_sample(size=100)*10 # 100 numbers b/w [0 and 10) 
y = randint(0,2, size=100)
```

```python
#create pandas dataframe for this data
df = pd.DataFrame(columns=['X1','X2','X3','Y'], data=list(zip(x1,x2,x3,y)))
df.head()
```

### CSV Format

```python
#save data into csv format
df.to_csv('random.csv')
```

### RecordIO Format

    -Supports Integer (int32) and Float (float32 and float64)
    -Recommended for large dataset
    -Non-numeric values to be converted to numeric values first

```python
df.head()
```

```python
X = df[['X1','X2','X3']].to_numpy()
y = df[['Y']].to_numpy(dtype='float64') #RecordIO does not support int64
y=y.ravel()
```

```python
print ('Shape of input variables is ',X.shape)
print ('Shape of target variables is ',y.shape)
```

```python
def write_recordio_file(filename, X,y):
    with open(filename,'wb') as f:
        smac.write_numpy_to_dense_tensor(f,X,y)
        
def read_recordio_file_as_json(filename):
    with open(filename, 'rb') as f:
        records = smac.read_records(f)
        for i,r in enumerate (records):
            print (r)
            
def read_recordio_file_as_numpy(filename):
    with open(filename, 'rb') as f:
        records = smac.read_records(f)
        for i,r in enumerate (records):
            print (r)            
```

```python
write_recordio_file('random.recordio',X,y)
```

```python
read_recordio_file('random.recordio')
```
